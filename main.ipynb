Present chinese postman problem examples

```python
cpp = ChinesePostman()
cpp.add_node('A', 'B', 3)
cpp.add_node('B', 'C', 4)
cpp.add_node('C', 'D', 5)
cpp.add_node('D', 'A', 2)
cpp.add_node('A', 'C', 1)

total_cost, path = cpp.solve()
print(f"\nOptimal Route:")
print(f"Path: {path}")
print(f"Total cost: {total_cost}\n")
cpp.plot_route()
```
